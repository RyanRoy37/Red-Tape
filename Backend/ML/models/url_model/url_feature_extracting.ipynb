```python
import numpy as np

# Select only numeric columns from the dataframe.
# df may contain strings (like 'benign'/'phishing') which cannot be used in correlation calculations.
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute the absolute correlation matrix between all numeric features.
# .corr() calculates Pearson correlation; .abs() converts negative values to positive for easier comparison.
corr_matrix = numeric_df.corr().abs()

# Create a mask for the upper triangle of the correlation matrix.
# This prevents duplicate comparisons (A-B and B-A) and avoids comparing a feature with itself.
upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

# Identify columns (features) to drop based on high multicollinearity.
# If any feature has correlation > 0.9 with another feature, it is considered redundant and marked for removal.
to_drop = [
    column for column in upper.columns 
    if any(upper[column] > 0.9)
]

# Display the list of redundant features with correlation > 0.9.
to_drop
```

reduce noise impact

```python
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel
from tqdm import tqdm

MODEL_NAME = "prajjwal1/bert-tiny"
BATCH_SIZE = 64          # can push higher on CPU
DEVICE = "cpu"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
bert_model = AutoModel.from_pretrained(MODEL_NAME)
bert_model.eval()


def mean_pooling(token_embeddings, attention_mask):
    mask = attention_mask.unsqueeze(-1).float()
    return (token_embeddings * mask).sum(1) / mask.sum(1)


def embed_urls(urls):
    all_embeddings = []

    with torch.no_grad():
        for i in tqdm(range(0, len(urls), BATCH_SIZE), desc="Embedding URLs"):
            batch = urls[i:i+BATCH_SIZE].tolist()

            encoded = tokenizer(
                batch,
                padding=True,
                truncation=True,
                max_length=64,   # IMPORTANT: keep small
                return_tensors="pt"
            )

            outputs = bert_model(**encoded)
            emb = mean_pooling(outputs.last_hidden_state, encoded["attention_mask"])
            all_embeddings.append(emb.numpy())
    tokenizer.save_pretrained('bert-tiny')
    bert_model.save_pretrained('bert-tiny')

    return np.vstack(all_embeddings)
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
    "Random Forest": RandomForestClassifier(
        n_estimators=300,        # Number of trees in the forest. More trees → more stable predictions but slower.
        max_depth=20,         # Maximum depth of each tree. None = grow fully until pure or min samples reached.
        random_state=42   ,      # Seed for reproducibility of results.
        n_jobs=1
    ),
    
    "XGBoost": XGBClassifier(
        n_estimators=300,        # Number of boosting rounds (number of trees added sequentially).
        max_depth=8,             # Maximum depth of each decision tree. Higher = more complex patterns, risk of overfitting.
        learning_rate=0.1,       # Shrinks the contribution of each tree. Lower = slower learning but better generalization.
        subsample=0.8,           # Percentage of rows sampled for each tree. Prevents overfitting.
        colsample_bytree=0.8,    # Percentage of features used for each tree. Reduces correlation between trees.
        eval_metric='logloss',   # Metric XGBoost uses for evaluation. Logloss = good for binary classification.
        random_state=42,         # For reproducibility.
        tree_method="hist"       # Uses histogram-based algorithm for faster CPU/GPU training.
    ),
    
    "LightGBM": LGBMClassifier(
        n_estimators=300,        # Number of boosting iterations (trees).
        max_depth=-1,            # -1 means no limit on depth; tree can grow until stopping criteria.
        learning_rate=0.1,       # Step size shrinkage to prevent overfitting.
        subsample=0.8,           # Fraction of data used per tree (row sampling).
        colsample_bytree=0.8,    # Fraction of features used per tree (feature sampling).
        random_state=42          # Reproducible training.
    )
}
```

Our data has 822000, my system does not have the resources to train them for so long so, for non-LightGBM models, we will only train with a subset of the training and test sets.

```python
from sklearn.model_selection import train_test_split

X_small, _, y_small, _ = train_test_split(
    X_train, y_train,
    train_size=100_000,
    random_state=42,
    stratify=y_train
)
```

```python
results = {}
import lightgbm as lgb

for name, model in models.items():
    print(f"\n==============================")
    print(f"TRAINING: {name}")
    print(f"==============================")

    # -------------------------------
    # LightGBM → full dataset
    # -------------------------------
    if isinstance(model, lgb.LGBMClassifier):
        print("Using FULL dataset for LightGBM")

        model.fit(
            X_train, y_train,
            eval_set=[(X_test, y_test)],
            eval_metric="multi_logloss",
            early_stopping_rounds=50,
            verbose=10
        )

        y_pred = model.predict(X_test)

    # --------------------------------
    # Other models → subset
    # --------------------------------
    else:
        print("Using SUBSET for non-LightGBM models")
        print(f"Subset size: {len(X_small)}")

        model.fit(X_small, y_small)
        y_pred = model.predict(X_test)

    # -------------------------------
    # Evaluation
    # -------------------------------
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc

    print(f"\nAccuracy ({name}): {acc:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
```

```python
comparison_df = pd.DataFrame.from_dict(results, orient='index', columns=["Accuracy"])
print(comparison_df.sort_values("Accuracy", ascending=False))
```
