```python
from scipy.stats import pointbiserialr
target=df["URL_Type_obf_Type"].map({"benign": 0, "phishing": 1})
corrs={}

for col in df.select_dtypes(include=['int64', 'float64']).columns:
    corrs[col]=pointbiserialr(df[col], target)[0]

import pandas as pd
import numpy as np

corr_df = pd.DataFrame.from_dict(corrs, orient='index', columns=['correlation'])
corr_df =corr_df.dropna()
threshold = 0.1
filtered_corr_df = corr_df[corr_df['correlation'].abs() >= threshold]
filtered_corr_df
```

```python
import numpy as np

# Select only numeric columns from the dataframe.
# df may contain strings (like 'benign'/'phishing') which cannot be used in correlation calculations.
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute the absolute correlation matrix between all numeric features.
# .corr() calculates Pearson correlation; .abs() converts negative values to positive for easier comparison.
corr_matrix = numeric_df.corr().abs()

# Create a mask for the upper triangle of the correlation matrix.
# This prevents duplicate comparisons (A-B and B-A) and avoids comparing a feature with itself.
upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

# Identify columns (features) to drop based on high multicollinearity.
# If any feature has correlation > 0.9 with another feature, it is considered redundant and marked for removal.
to_drop = [
    column for column in upper.columns 
    if any(upper[column] > 0.9)
]

# Display the list of redundant features with correlation > 0.9.
to_drop
```

We will not be dropping the columns because, the model we are using is Tree-based.
Tree models automatically:
ignore redundant features
select strong splits
reduce noise impact

```python
#Seperating features into dependent and independent sets
x = df.drop(columns=["URL_Type_obf_Type"])
y=df['URL_Type_obf_Type']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
# Replace infinite values with NaN
X_train = X_train.replace([np.inf, -np.inf], np.nan)
X_test = X_test.replace([np.inf, -np.inf], np.nan)
# Replace NaN with median (or mean)
X_train = X_train.fillna(X_train.median())
X_test = X_test.fillna(X_train.median())
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
    "Random Forest": RandomForestClassifier(
        n_estimators=300,        # Number of trees in the forest. More trees → more stable predictions but slower.
        max_depth=None,         # Maximum depth of each tree. None = grow fully until pure or min samples reached.
        random_state=42         # Seed for reproducibility of results.
    ),
    
    "XGBoost": XGBClassifier(
        n_estimators=300,        # Number of boosting rounds (number of trees added sequentially).
        max_depth=8,             # Maximum depth of each decision tree. Higher = more complex patterns, risk of overfitting.
        learning_rate=0.1,       # Shrinks the contribution of each tree. Lower = slower learning but better generalization.
        subsample=0.8,           # Percentage of rows sampled for each tree. Prevents overfitting.
        colsample_bytree=0.8,    # Percentage of features used for each tree. Reduces correlation between trees.
        eval_metric='logloss',   # Metric XGBoost uses for evaluation. Logloss = good for binary classification.
        random_state=42,         # For reproducibility.
        tree_method="hist"       # Uses histogram-based algorithm for faster CPU/GPU training.
    ),
    
    "LightGBM": LGBMClassifier(
        n_estimators=300,        # Number of boosting iterations (trees).
        max_depth=-1,            # -1 means no limit on depth; tree can grow until stopping criteria.
        learning_rate=0.1,       # Step size shrinkage to prevent overfitting.
        subsample=0.8,           # Fraction of data used per tree (row sampling).
        colsample_bytree=0.8,    # Fraction of features used per tree (feature sampling).
        random_state=42          # Reproducible training.
    )
}
results = {}
for name, model in models.items():
    print(f"TRAINING: {name}")
    # Train
    model.fit(X_train, y_train)
    # Predict
    y_pred = model.predict(X_test)
    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    # Store results
    results[name] = acc
    # Print report
    print(f"Accuracy: {acc:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
```

```python
comparison_df = pd.DataFrame.from_dict(results, orient='index', columns=["Accuracy"])
print(comparison_df.sort_values("Accuracy", ascending=False))
```
